```python
%%sql
SELECT s.City, s.CompanyName, p.ProductName, p.UnitPrice
FROM suppliers s INNER JOIN products p
                    ON s.ID = p.SupplierID
WHERE s.country = "Japan"
ORDER BY s.City
```

```python
%%sql
SELECT s.Country, AVG(p.UnitPrice) AS avg_unitprice, COUNT(p.ProductName) AS count_product, Count(DISTINCT s.City) AS count_city
FROM suppliers s INNER JOIN products p
                  ON s.ID = p.SupplierID
GROUP BY s.Country
ORDER BY avg_unitprice DESC, count_product;
```

```python
%%sql
SELECT c.firstname, c.lastname, o.ordernumber, o.orderdate, o.totalamount
FROM orders o INNER JOIN customers c ON o.CustomerId = c.Id
WHERE o.OrderDate > 20140229 AND o.OrderDate < 20140401
ORDER BY o.orderdate DESC, o.totalamount DESC
```

```python
%%sql
SELECT year(orderdate) AS order_year, p.productname, MIN(o.totalamount)
FROM orders o INNER JOIN orderitems od ON o.Id = od.OrderId
                INNER JOIN products p ON od.ProductId = p.Id
GROUP BY order_year
```

```python
%%sql
SELECT o.ordernumber, o.orderdate, CONCAT(c.firstname, ' ', c.lastname) AS customer_name, totalamount
FROM orders o INNER JOIN customers c ON o.CustomerId = c.Id
WHERE DAYNAME(o.orderdate) = 'Sunday'
ORDER BY o.orderdate DESC
```

```python
%%sql
SELECT CONCAT(c.firstname, ' ', c.lastname) AS customer_name, c.country, c.city,  COUNT(o.id) AS count_orders
FROM orders o INNER JOIN customers c ON o.CustomerId = c.Id
GROUP BY customer_name
HAVING count_orders > 15
ORDER BY count_orders DESC 
```

```python
%%sql
SELECT o.Id, o.orderdate, o.ordernumber, o.CustomerId, o.totalamount
FROM orders o INNER JOIN customers c ON o.CustomerId = c.Id
            INNER JOIN orderitems od ON o.Id = od.OrderId
            INNER JOIN products p ON od.ProductId = p.Id
WHERE p.ProductName = 'Chai'
ORDER BY o.totalamount DESC 
```

```python
%%sql
SELECT p.ProductName, SUM(i.Quantity) AS Sum_quantity
FROM products p LEFT JOIN orderitems i ON p.id = i.productid 
                LEFT JOIN orders o ON o.Id = i.OrderId
WHERE p.IsDiscontinued = 0 
GROUP BY p.ProductName
ORDER BY sum_quantity
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = %sql SELECT CONCAT(c.firstname, ' ', c.lastname) AS customer_name, COUNT(o.id) AS count_orders FROM customers c INNER JOIN orders o ON c.id = o.CustomerId  GROUP BY customer_name ORDER BY count_orders DESC LIMIT 10
data
```

```python
df = data.DataFrame()
df
```

```python
sns.barplot(data = df, y = 'customer_name', x = 'count_orders')
plt.xlabel('Số đơn đặt hàng')
plt.ylabel('Tên khách hàng')
plt.title('10 Khách hàng có nhiều đơn đặt hàng nhất')
plt.show()
```

```python
data1 = %sql SELECT year(orderdate) AS year, sum(totalamount) AS total_amount, COUNT(*) AS count FROM orders GROUP BY year(orderdate)
data1
```
